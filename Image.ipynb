```python
from tkinter import *
from PIL import ImageTk, Image            # PIL is pillow

mainWindow = Tk()

image1 = Image.open("C:/Users/tuhin/jupyter notebook/practice/picture/tuhin.png")
image2 = Image.open("C:/Users/tuhin/jupyter notebook/practice/picture/alomgir.png")
image3 = Image.open("C:/Users/tuhin/jupyter notebook/practice/picture/sumon.png")
image4 = Image.open("C:/Users/tuhin/jupyter notebook/practice/picture/jesan.png")
image5 = Image.open("C:/Users/tuhin/jupyter notebook/practice/picture/nazrul.png")
image6 = Image.open("C:/Users/tuhin/jupyter notebook/practice/picture/halim.png")

image1 = image1.resize((250, 300), Image.ANTIALIAS)     # (250, 250) is (width, height)
image2 = image2.resize((250, 300), Image.ANTIALIAS)
image3 = image3.resize((250, 300), Image.ANTIALIAS)
image4 = image4.resize((250, 300), Image.ANTIALIAS)
image5 = image5.resize((250, 300), Image.ANTIALIAS)
image6 = image6.resize((250, 300), Image.ANTIALIAS)

my_img1 = ImageTk.PhotoImage(image1)
my_img2 = ImageTk.PhotoImage(image2)
my_img3 = ImageTk.PhotoImage(image3)
my_img4 = ImageTk.PhotoImage(image4)
my_img5 = ImageTk.PhotoImage(image5)
my_img6 = ImageTk.PhotoImage(image6)

image_list = [my_img1, my_img2, my_img3, my_img4, my_img5, my_img6]

my_label = Label(image=my_img1)
my_label.grid(row=0, column=0, columnspan=3)

status = Label(mainWindow, text="Image 1 of " + str(len(image_list)), bd=2, relief=SUNKEN, anchor=E)

def forward(image_number):
    global my_label
    global button_forward
    global button_back

    my_label.grid_forget()             # remove the previous image
    button_forward.grid_forget()
    button_back.grid_forget()

    my_label = Label(image= image_list[image_number])
    button_forward = Button(mainWindow, text=">>", command=lambda: forward(image_number+1))
    button_back = Button(mainWindow, text="<<", command=lambda: back(image_number-1))

    if image_number == len(image_list) -1:
        button_forward = Button(mainWindow, text=">>", state=DISABLED)

    my_label.grid(row=0, column=0, columnspan=3)
    button_back.grid(row=1, column=0, pady=10)
    button_forward.grid(row=1, column=2)

    status = Label(mainWindow, text="Image " + str(image_number+1) + " of " + str(len(image_list)), bd=2, relief=SUNKEN, anchor=E)
    status.grid(row=2, column=0, columnspan=3, sticky=W + E)

def back(image_number):
    global my_label
    global button_forward
    global button_back

    my_label.grid_forget()        # remove the previous image
    button_forward.grid_forget()
    button_back.grid_forget()

    my_label = Label(image=image_list[image_number])
    button_forward = Button(mainWindow, text=">>", command=lambda: forward(image_number + 1))
    button_back = Button(mainWindow, text="<<", command=lambda: back(image_number - 1))

    if image_number == 0:
        button_back = Button(mainWindow, text="<<", state=DISABLED)

    my_label.grid(row=0, column=0, columnspan=3, pady=10)
    button_back.grid(row=1, column=0)
    button_forward.grid(row=1, column=2)

    status = Label(mainWindow, text="Image " + str(image_number + 1) + " of " + str(len(image_list)), bd=2,
                   relief=SUNKEN, anchor=E)
    status.grid(row=2, column=0, columnspan=3, sticky=W + E)

button_back = Button(mainWindow, text="<<", state=DISABLED, command=back)
button_exit = Button(mainWindow, text="Exit Program", command=mainWindow.quit)
button_forward = Button(mainWindow, text=">>", command=lambda: forward(1))

button_back.grid(row=1, column=0)
button_exit.grid(row=1, column=1)
button_forward.grid(row=1, column=2, pady=10)

status.grid(row=2, column=0, columnspan=3, sticky=W+E)

mainWindow.mainloop()
```
