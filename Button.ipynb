## Basic

```python
from tkinter import *

mainWindow = Tk()

myButton1 = Button(mainWindow, text="Click Me!")
myButton2 = Button(mainWindow, text="Click Me!", state=DISABLED)     # state disable the button
myButton3 = Button(mainWindow, text="Click Me!", padx=50)            # padx = padding in x axis
myButton4 = Button(mainWindow, text="Click Me!", padx=50, pady=50)   # pady = padding in y axis

myButton1.pack()
myButton2.pack()
myButton3.pack()
myButton4.pack()

mainWindow.mainloop()
```

## Exit Button

## Button with Action

```python
from tkinter import *

mainWindow = Tk()

def myClick():
    myLabel = Label(mainWindow, text='Look! I clicked a Button')
    myLabel.pack()

myButton = Button(mainWindow, text="Click Me!", command=myClick)
myButton.pack()

mainWindow.mainloop()
```

```python
from tkinter import *

mainWindow = Tk()

def myClick():
    myLabel = Label(mainWindow, text=myTextField.get())
    myLabel.pack()

myTextField = Entry(mainWindow, width=80, bg="aqua", fg="red", borderwidth=5)
myTextField.pack()

myButton = Button(mainWindow, text="Show My Name", command=myClick)
myButton.pack()

mainWindow.mainloop()
```
