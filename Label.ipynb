```python
# anchor ==> where the text is positioned if the widget has more space than the text needs.
        # N, NE, E, SE, S, SW, W, NW or default is CENTER
# fg ==> Foreground color or text color.
# bg ==> Background color
# borderwidth ==> the width of the label border. or used bd for borderwidth
# font ==> font of the label text
# height ==> height of the label
# width ==> width of the label
# state ==> controls how the label is rendered. ACTIVE or DISABLED
# justify ==> Defines how to align multiple lines of text. LEFT, RIGHT, CENTER (text alignment)
# relief ==> Border decoration. FLAT, SUNKEN, RAISED, GROOVE, and RIDGE.
```

```python
from tkinter import *
from PIL import ImageTk, Image

mainWindow = Tk()
mainWindow.geometry('1000x800+300+100')

myLabel1 = Label(mainWindow,
                text="This is my Label",
                fg='red',
                bg='blue',
                font=("Helvetica", 20, "italic"),
                width=150,
                height=3,
                anchor=E,
                justify=LEFT,
                padx=5,
                pady=5,
                borderwidth=10,
                relief=FLAT)
myLabel1.pack(padx=5, pady=10)

myLabel2 = Label(mainWindow,
                text="This is my Label",
                fg='red',
                bg='blue',
                font=("Times", 16),
                width=40,
                height=3,
                anchor=N,
                justify=RIGHT,
                padx=5,
                pady=5,
                borderwidth=10,
                relief=SUNKEN)
myLabel2.pack(padx=5, pady=5)

myLabel3 = Label(mainWindow,
                text="This is my Label",
                fg='red',
                bg='blue',
                font=("Gotham", 30),
                width=150,
                height=3,
                anchor=W,
                justify=CENTER,
                padx=5,
                pady=5,
                borderwidth=10,
                relief=GROOVE)
myLabel3.pack(padx=5, pady=10)

myLabel4 = Label(mainWindow,
                text="This is my Label",
                fg='red',
                bg='blue',
                font=("Arial", 10),
                width=50,
                height=7,
                anchor=S,
                justify=LEFT,
                padx=5,
                pady=5,
                borderwidth=10,
                relief=RIDGE)
myLabel4.pack(padx=5, pady=5)

image1 = Image.open("C:/Users/tuhin/jupyter notebook/practice/picture/tuhin.png")
image1 = image1.resize((100, 150), Image.ANTIALIAS)
my_img1 = ImageTk.PhotoImage(image1)
my_label = Label(image=my_img1,
                borderwidth=10,
                relief=RAISED)
my_label.pack()

mainWindow.mainloop()
```
