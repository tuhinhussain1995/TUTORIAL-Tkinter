```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')
mainWindow.configure(padx=10,pady=100)

def select():
    sel = "Value = " + str(value.get())
    label.config(text=sel)

value = DoubleVar()

spin = Spinbox(mainWindow,
               from_=0,
               to=1000,
               font='Times 20 bold',
               fg='red',
               bg='orange',
               activebackground='yellow',
               width=20,
               textvariable=value,
               cursor='arrow',
               justify=CENTER,
               highlightbackground='pink',
               highlightcolor='aqua',
               highlightthickness=10,
               bd=10,
               relief=SUNKEN,
               buttonbackground='green',
               buttoncursor='cross',
               increment=10
               )
spin.pack()

btn = Button(mainWindow, text="Value", command=select)
btn.pack(anchor=CENTER)

label = Label(mainWindow)
label.pack()

mainWindow.mainloop()
```
