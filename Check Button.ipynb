```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

checkvar1 = IntVar()
checkvar2 = IntVar()
checkvar3 = IntVar()

def called1():
    myVal = checkvar1.get()
    print(myVal)
    myLabel = Label(mainWindow, text="Python is Called", font="Black 12 italic", pady=20)
    myLabel.pack()

def called2():
    myVal = checkvar3.get()
    print(myVal)
    myLabel = Label(mainWindow, text="Java is Called", font="Black 12 italic", pady=20)
    myLabel.pack()

chkbtn1 = Checkbutton(mainWindow,
                      text="Python",
                      variable=checkvar1,
                      onvalue=1,
                      offvalue=0,
                      font="Times 20 bold",
                      fg='red',
                      bg='green',
                      bd=5,
                      relief=RAISED,
                      command=called1,
                      height=2,
                      width=10,
                      padx=10,
                      pady=10,
                      justify=RIGHT,)

chkbtn2 = Checkbutton(mainWindow,
                      text="C++",
                      variable=checkvar2,
                      onvalue=1,
                      offvalue=0,
                      font="Times 20 bold",
                      fg='red',
                      bg='gray',
                      bd=5,
                      relief=RAISED,
                      command=called1,
                      height=2,
                      width=10,
                      padx=10,
                      pady=10,
                      justify=RIGHT,
                      state=DISABLED)

chkbtn3 = Checkbutton(mainWindow,
                      text="Java",
                      variable=checkvar3,
                      onvalue=1,
                      offvalue=0,
                      font="Times 20 bold",
                      fg='blue',
                      bg='yellow',
                      bd=5,
                      relief=RAISED,
                      command=called2,
                      height=2,
                      width=10,
                      padx=10,
                      pady=10,
                      justify=RIGHT,)

chkbtn1.pack()
chkbtn2.pack()
chkbtn3.pack()

mainWindow.mainloop()
```
