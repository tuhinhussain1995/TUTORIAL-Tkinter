```python
from tkinter import *

mainWindow = Tk()
mainWindow.title("Calculator")
mainWindow.iconbitmap('C:/Users/tuhin/Python Masterclass/hnet.com-image.ico')

myLabelTop = Label(mainWindow, text="Standard Calculator", font=("Arial", 16), pady=20)
myLabelTop.grid(row=0, column=0, columnspan=4)

textField = Entry(mainWindow, width=16, borderwidth=5, justify=RIGHT, font=("Arial Bold", 25))
textField.grid(row=1, column=0, columnspan=4, padx=10, pady=20)

def button_click(number):
    current = textField.get()
    textField.delete(0, END)
    textField.insert(0, (str(current) + str(number)))

def button_clear():
    textField.delete(0, END)

def button_add():
    first_num = textField.get()
    global num1, calculation
    calculation = "addition"
    num1 = float(first_num)
    textField.delete(0, END)

def button_subtract():
    first_num = textField.get()
    global num1, calculation
    calculation = "subtraction"
    num1 = float(first_num)
    textField.delete(0, END)

def button_multiply():
    first_num = textField.get()
    global num1, calculation
    calculation = "multiplication"
    num1 = float(first_num)
    textField.delete(0, END)

def button_divide():
    first_num = textField.get()
    global num1, calculation
    calculation = "division"
    num1 = float(first_num)
    textField.delete(0, END)

def button_remainder():
    first_num = textField.get()
    global num1, calculation
    calculation = "remainder"
    num1 = float(first_num)
    textField.delete(0, END)

def button_square():
    first_num = textField.get()
    global num1, calculation
    calculation = "square"
    num1 = float(first_num)
    textField.delete(0, END)
    result = '{0:.2f}'.format(pow(num1, 2))
    textField.insert(0, result)

def button_equal():
    second_num = textField.get()
    textField.delete(0, END)

    if calculation == 'addition':
        result = '{0:.2f}'.format(num1 + float(second_num))
        textField.insert(0, result)
    elif calculation == 'subtraction':
        result = '{0:.2f}'.format(num1 - float(second_num))
        textField.insert(0, result)
    elif calculation == 'multiplication':
        result = '{0:.2f}'.format(num1 * float(second_num))
        textField.insert(0, result)
    elif calculation == 'division':
        result = '{0:.2f}'.format(num1 / float(second_num))
        textField.insert(0, result)
    elif calculation == 'remainder':
        result = '{0:.2f}'.format(num1 % float(second_num))
        textField.insert(0, result)

def button_cut():
    first_num = textField.get()
    textField.delete(0, END)
    first_num = first_num[0:(len(first_num))-1]
    textField.insert(0, first_num)

        #if we want to pass anything, we must need to use lambda
buttonl = Button(mainWindow, text="1", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(1))
button2 = Button(mainWindow, text="2", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(2))
button3 = Button(mainWindow, text="3", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(3))
button4 = Button(mainWindow, text="4", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(4))
button5 = Button(mainWindow, text="5", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(5))
button6 = Button(mainWindow, text="6", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(6))
button7 = Button(mainWindow, text="7", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(7))
button8 = Button(mainWindow, text="8", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(8))
button9 = Button(mainWindow, text="9", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(9))
button0 = Button(mainWindow, text="0", padx=26, pady=8, font = ("Arial Bold", 14), command= lambda: button_click(0))
button_dot = Button(mainWindow, text=".", padx=28, pady=8, bg="#D6D6D6", font = ("Arial Bold", 14), command= lambda: button_click('.'))

button_remainder = Button(mainWindow, text="%", padx=26, pady=11, bg="#D6D6D6", font = ("Arial", 11), command= button_remainder)
button_square = Button(mainWindow, text="X\u00B2", padx=26, pady=11, bg="#D6D6D6", font = ("Arial", 11), command= button_square)
button_clear = Button(mainWindow, text="Clear", padx=14, pady=11, bg="#D6D6D6", font = ("Arial", 11), command= button_clear)
button_cut = Button(mainWindow, text="\u232B", padx=23, pady=11, bg="#D6D6D6", font = ("Arial", 11), command= button_cut)

button_add = Button(mainWindow, text="+", padx=26, pady=8, bg="#D6D6D6", font = ("Arial", 14), command= button_add)
button_subtract = Button(mainWindow, text="\u2212", padx=26, pady=8, bg="#D6D6D6", font = ("Arial", 14), command= button_subtract)
button_multiply = Button(mainWindow, text="\u00D7", padx=26, pady=8, bg="#D6D6D6", font = ("Arial", 14), command= button_multiply)
button_divide = Button(mainWindow, text="\u00F7", padx=26, pady=8, bg="#D6D6D6", font = ("Arial", 14), command= button_divide)
button_equal = Button(mainWindow, text="=", padx=26, pady=8, bg="#8ABAE0", font = ("Arial", 14), command= button_equal)

button_remainder.grid(row=2, column=0)
button_square.grid(row=2, column=1)
button_clear.grid(row=2, column=2)
button_cut.grid(row=2, column=3)

button7.grid(row=3, column=0)
button8.grid(row=3, column=1)
button9.grid(row=3, column=2)
button_divide.grid(row=3, column=3)

button4.grid(row=4, column=0)
button5.grid(row=4, column=1)
button6.grid(row=4, column=2)
button_multiply.grid(row=4, column=3)

buttonl.grid(row=5, column=0)
button2.grid(row=5, column=1)
button3.grid(row=5, column=2)
button_add.grid(row=5, column=3)

button_dot.grid(row=6, column=0)
button0.grid(row=6, column=1)
button_subtract.grid(row=6, column=2)
button_equal.grid(row=6, column=3)

mainWindow.mainloop()
```
