```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

lbl = Label(text="Favourite programming language:")
lbl.pack()

radio = IntVar()

def selection():
   selection = "You selected the option " + str(radio.get())
   label.config(text = selection, font='Times 20 bold', fg='green')



R1 = Radiobutton(mainWindow,
                 text="C",
                 variable=radio,
                 value=1,
                 command=selection,
                 font='Arial 20 bold',
                 bg='yellow',
                 fg='blue',
                 activebackground='green',
                 activeforeground='red',
                 anchor=W,
                 width=10,
                 height=2,
                 justify=RIGHT,
                 bd=5,
                 relief=RAISED,
                 padx=10,
                 pady=10)
R1.pack(anchor=W)

R2 = Radiobutton(mainWindow,
                 text="C++",
                 variable=radio,
                 value=2,
                 command=selection,
                 font='Arial 20 bold',
                 bg='yellow',
                 fg='blue',
                 activebackground='green',
                 activeforeground='red',
                 anchor=W,
                 width=10,
                 height=2,
                 justify=RIGHT,
                 bd=5,
                 relief=RAISED,
                 padx=10,
                 pady=10)
R2.pack(anchor=W)

R3 = Radiobutton(mainWindow,
                 text="Java",
                 variable=radio,
                 value=3,
                 command=selection,
                 font='Arial 20 bold',
                 bg='yellow',
                 fg='blue',
                 activebackground='green',
                 activeforeground='red',
                 anchor=W,
                 width=10,
                 height=2,
                 justify=RIGHT,
                 bd=5,
                 relief=RAISED,
                 padx=10,
                 pady=10)
R3.pack(anchor=W)

label = Label(mainWindow)
label.pack()

mainWindow.mainloop()
```
