```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

def open():
    top = Toplevel(mainWindow,
                   bg='yellow',
                   bd=10,
                   relief=RAISED,
                   cursor='clock',
                   highlightbackground='blue',
                   highlightcolor='orange',
                   highlightthickness=10,
                   padx=10,
                   pady=10,)
    top.geometry('330x290+400+170')

    def exitTopLevel():
        top.destroy()

    button_quit = Button(top, text="Exit Program", command=exitTopLevel)
    button_quit.pack()
    top.mainloop()


btn = Button(mainWindow, text="open", command=open)

btn.place(x=75, y=50)

mainWindow.mainloop()
```

```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

def open():
    menubar = Menu(mainWindow)

    top = Toplevel(mainWindow, menu=menubar,
                   bg='yellow',
                   bd=10,
                   relief=RAISED,
                   cursor='clock',
                   highlightbackground='blue',
                   highlightcolor='orange',
                   highlightthickness=10,
                   padx=10,
                   pady=10,)
    top.geometry('330x290+400+170')

    file = Menu(menubar,
                tearoff=0,
                font='Times 15 bold italic',
                fg='red',
                bg='yellow',
                activebackground='orange',
                activeforeground='white',
                relief=RAISED)
    file.add_command(label="New")
    file.add_command(label="Open")
    file.add_command(label="Save")
    file.add_command(label="Save as...")
    file.add_command(label="Close")
    file.add_separator()
    file.add_command(label="Exit", command=mainWindow.quit)

    menubar.add_cascade(label="File", menu=file)

    def exitTopLevel():
        top.destroy()

    button_quit = Button(top, text="Exit Program", command=exitTopLevel)
    button_quit.pack()
    top.mainloop()


btn = Button(mainWindow, text="open", command=open)

btn.place(x=75, y=50)

mainWindow.mainloop()
```
