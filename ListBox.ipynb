```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

myLabel = Label(mainWindow, text="A list of favourite countries...")

listbox = Listbox(mainWindow,
                  bg='gray',
                  fg='red',
                  font="Times 14 bold italic",
                  cursor="star",
                  width=20,
                  height=10,
                  highlightcolor='orange',
                  highlightthickness=5,
                  selectbackground="green",
                  bd=5,
                  relief=RAISED)

listbox.insert(1, "India")
listbox.insert(2, "USA")
listbox.insert(3, "Japan")
listbox.insert(4, "Austrelia")

myLabel.pack()
listbox.pack()

dltBtn = Button(mainWindow, text = "Delete", command = lambda listbox=listbox: listbox.delete(ANCHOR))
dltBtn.pack()

mainWindow.mainloop()
```

```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

def deleteSelected():
    listbox.delete(ANCHOR)

def deleteAll():
    listbox.delete(0, END)

myLabel = Label(mainWindow, text="A list of favourite countries...")

listbox = Listbox(mainWindow,
                  bg='gray',
                  fg='red',
                  font="Times 14 bold italic",
                  cursor="star",
                  width=20,
                  height=10,
                  highlightcolor='orange',
                  highlightthickness=5,
                  selectbackground="green",
                  bd=5,
                  relief=RAISED)

listbox.insert(0, "Malaysia")
listbox.insert(1, "India")
listbox.insert(2, "USA")
listbox.insert(3, "Japan")
listbox.insert(4, "Austrelia")

myLabel.pack()
listbox.pack()

listbox.delete(2)              # delete only one item using index number

dltBtn1 = Button(mainWindow, text = "Delete Selected", command = deleteSelected)
dltBtn1.pack()

dltBtn2 = Button(mainWindow, text = "Delete All", command = deleteAll)
dltBtn2.pack()

mainWindow.mainloop()
```
