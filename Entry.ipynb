```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

number1 = StringVar()
number2 = StringVar()

def result():
    num1 = number1.get()
    num2 = number2.get()
    add = int(num1) + int(num2)

    myLabel = Label(mainWindow, text="{0}+{1} = {2}".format(num1, num2, add))
    myLabel.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

label1 = Label(mainWindow, text="A")
label2 = Label(mainWindow, text="B")
label1.grid(row=0, column=0)
label2.grid(row=1, column=0)

entry1 = Entry(mainWindow, textvariable=number1)
entry2 = Entry(mainWindow, textvariable=number2)
entry1.grid(row=0, column=1)
entry2.grid(row=1, column=1)

sbmitbtn = Button(mainWindow, text="Result", activebackground="pink", activeforeground="blue", command=result)
sbmitbtn.grid(row=2, column=0, columnspan=2)

mainWindow.mainloop()
```

```python
from tkinter import *

mainWindow = Tk()
mainWindow.geometry('600x500+300+100')

def replay():
    n = myName.get()
    e = myEmail.get()
    p = myPassword.get()

    myLabel1 = Label(mainWindow, text=n)
    myLabel2 = Label(mainWindow, text=e)
    myLabel3 = Label(mainWindow, text=p)

    myLabel1.grid(row=4, column=0, columnspan=2)
    myLabel2.grid(row=5, column=0, columnspan=2)
    myLabel3.grid(row=6, column=0, columnspan=2)

myName = StringVar()
myEmail = StringVar()
myPassword = StringVar()

name = Label(mainWindow, text="Name")
email = Label(mainWindow, text="Email")
password = Label(mainWindow, text="Password")

e1 = Entry(mainWindow,
           bg='gray',
           font='Arial 18 bold italic',
           fg="aqua",
           highlightthickness=5,
           highlightbackground="blue",
           justify=RIGHT,
           bd=5,
           relief=SUNKEN,
           textvariable=myName,
           cursor="plus")
e2 = Entry(mainWindow,
           bg='gray',
           font='Arial 18 bold italic',
           fg="aqua",
           highlightthickness=5,
           highlightbackground="green",
           justify=RIGHT,
           bd=5,
           relief=SUNKEN,
           textvariable=myEmail,
           cursor="circle")
e3 = Entry(mainWindow,
           bg='gray',
           font='Arial 18 bold italic',
           fg="aqua",
           highlightthickness=5,
           highlightbackground="orange",
           justify=RIGHT,
           bd=5,
           relief=SUNKEN,
           textvariable=myPassword,
           show='*',
           cursor="arrow")

sbmitbtn = Button(mainWindow, text="Submit", activebackground="pink", activeforeground="blue", command=replay)

name.grid(row=0, column=0)
email.grid(row=1, column=0)
password.grid(row=2, column=0)
e1.grid(row=0, column=1)
e2.grid(row=1, column=1)
e3.grid(row=2, column=1)
sbmitbtn.grid(row=3, column=0, columnspan=2)

mainWindow.mainloop()
```
